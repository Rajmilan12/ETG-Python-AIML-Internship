```python
import json

def menuDisplay():
    print('\n\n\n=================================')
    print('-: Inventory Management Menu :-')
    print('=================================')
    print('(1) Add New Item to The Inventory')
    print('(2) Remove Item from The Inventory')
    print('(3) Update The Inventory')
    print('(4) Search Item in The Inventory')
    print('(5) Print Inventory Report')
    print('(0) Quit')
    CHOICE = int(input("Enter choice: "))
    menuSelection(CHOICE)

def menuSelection(CHOICE):
    if CHOICE == 1:
        addInventory()
    elif CHOICE == 2:
        removeInventory()
    elif CHOICE == 3:
        updateInventory()
    elif CHOICE == 4:
        searchInventory()
    elif CHOICE == 5:
        printInventory()
    elif CHOICE == 0:
        print("Exiting Now...Done")
        exit()

def addInventory():
    fd = open('product_inventory.json','r')
    txt = fd.read()
    fd.close()
    dct = json.loads(txt)
    print("Adding Inventory")
    print("===============================================")
    item_name = input("Enter the name of the item: ")
    flag=0
    for i in dct.keys():
      if item_name==dct[i]['Name']:
        print("Item already Exists")
        flag=1
        break
    
    if flag==0:
      item_price = input("Enter the price of the item: ")
      item_quantity = input("Enter the quantity of the item: ")
      item_discount = input("Enter the discount of the item: ")
      item_shelf_life = input("Enter the shelf life of the item: ")
      id  = int(max(dct.keys()))+1
      dct.update({id:{'Name':item_name,'Price':item_price,'Qty':item_quantity,'discount':item_discount,'ShelfLife':item_shelf_life}})
      print('{} added Successfully'.format(item_name))
      js = json.dumps(dct)
      fd = open('product_inventory.json','w')
      fd.write(js)
      fd.close()
    menuDisplay()
    
def removeInventory():
    print("Removing Inventory")
    print("==================")
    item_description = input("Enter the item name to remove from inventory: ")
    fd = open('product_inventory.json','r')
    txt = fd.read()
    fd.close()
    dct = json.loads(txt)
    flag=0
    for i in dct.keys():
      if dct[i]['Name']==item_description:
        flag=1;
        dct.pop(i)
        print('{} removed Successfully'.format(item_description))
        break
    if flag==0:
      print("Item does not Exist in Inventory..")
    js = json.dumps(dct)
    fd = open('product_inventory.json','w')
    fd.write(js)
    fd.close()
    menuDisplay()

def updateInventory():
    print("Updating Inventory")
    print("==================")
    item_description = input('Enter the item name to update: ')
    item_quantity = int(input("Enter the updated quantity to be replaced: "))

    fd = open('product_inventory.json','r')
    txt = fd.read()
    fd.close()
    dct = json.loads(txt)
    for i in dct.keys():
      if dct[i]['Name']==item_description:
        dct[i]['Qty'] = item_quantity
        break
    js = json.dumps(dct)
    fd = open('product_inventory.json','w')
    fd.write(js)
    fd.close()
    print('{} updated Successfully'.format(item_description))
    menuDisplay()

def searchInventory():
    print('Searching Inventory')
    print('===================')
    item_description = input('Enter the name of the item: ')
    fd = open('product_inventory.json','r')
    txt = fd.read()
    fd.close()
    dct = json.loads(txt)
    flag=0
    for i in dct.keys():
      if dct[i]['Name']==item_description:
        print(dct[i])
        flag=1
        break
    if flag==0:
      print("Item Not found in Inventory..")
    menuDisplay()
        
def printInventory():
    import pandas as pd
    fd = open('product_inventory.json','r')
    txt = fd.read()
    fd.close()
    dct = json.loads(txt)
    print('Current Inventory')
    print('-----------------')
    # print('Item\t\tPrice\t\tQuantity\t\tDiscount\t\tShelf_Life')
    # print('====\t\t=====\t\t========\t\t========\t\t==========')
    # for i in dct.keys():
    #    print(dct[i]['Name'],'\t',dct[i]['Price'],'\t',dct[i]['Qty'],'\t',dct[i]['discount'],'\t',dct[i]['ShelfLife'])
    ss = pd.DataFrame.from_dict(dct)
    display(ss)
    menuDisplay()

menuDisplay()
```
